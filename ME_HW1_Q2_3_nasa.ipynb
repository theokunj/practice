```python
import pandas as pd
df = pd.read_csv('/content/airfoil_self_noise.dat', header=None, delimiter='\t')
df.columns = ['Freq (Hz)', 'Angle of attack (deg)', 'Chord length, (m)', 'Free-stream velocity (m/s)', 'Suction side displacement thickness (m)', 'Scaled sound pressure level (dB)']
df.head()
```

```python
min_values = df.min()
max_values = df.max()
print("Minimum values per column:")
print(min_values)
print("\nMaximum values per column:")
print(max_values)
```

```python
data_range = max_values - min_values
print("Range per column:")
print(data_range)
```

```python
mean_values = df.mean()
print("Mean values per column:")
print(mean_values)
```

```python
std_values = df.std()
print("Standard deviation values per column:")
print(std_values)
```

```python
median_values = df.median()
print("Median values per column:")
print(median_values)
```

```python
kurtosis_values = df.kurtosis()
print("Kurtosis values per column:")
print(kurtosis_values)
```

```python
skew_values = df.skew()
print("Skewness values per column:")
print(skew_values)
```
